```python
# TASK 1 epoch counter

class EpochCounter:
    def __init__(self, total_epochs):
        self.total = total_epochs
        self.current = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.current < self.total:
            self.current += 1
            return f"Epoch {self.current}/{self.total}"
        else:
            raise StopIteration


# Usage
for epoch in EpochCounter(5):
    print(epoch)
```

```python
# task 2 using yield
def loss_simulator(start_loss, decay, epochs):
    for epoch in range(1, epochs + 1):
        loss = start_loss * (decay ** epoch)
        yield round(loss, 4)


# Usage
for loss in loss_simulator(1.0, 0.7, 6):
    print(loss)
```

```python
# task 3
def batch_generator(data, batch_size):
    for i in range(0, len(data), batch_size):
        yield data[i:i + batch_size]


# Usage
data = list(range(1, 21))

for batch in batch_generator(data, 4):
    print(batch)
```

```python
# task 4 memory comparison
import sys

# List comprehension
list_comp = [x**2 for x in range(100000)]

# Generator expression
gen_exp = (x**2 for x in range(100000))

print("List size:", sys.getsizeof(list_comp))
print("Generator size:", sys.getsizeof(gen_exp))
```
