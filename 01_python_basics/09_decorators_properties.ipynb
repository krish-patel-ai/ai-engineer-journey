# AI Engineer Python Decorators and Properties

```python
# task 1 decorator
def log_call(func):
    def wrapper(*args,**kwargs):
        print(f"Calling : {func.__name__}")
        result=func(*args,**kwargs)
        print(f"Finished : {func.__name__}")
        return result
    return wrapper
@log_call
def evaluate_model(name):
    print(f"Evaluating {name}")
evaluate_model("XGBoost")
```

```python
# task 2 using time.time
import time
def timer(func):
    def wrapper(*args,**kwargs):
        start=time.time()
        result=func(*args,**kwargs)
        end=time.time()
        print(f"time taken : {end-start} seconds")
        return result
    return wrapper
@timer
def train(epochs):
    time.sleep(0.05*epochs)
train(3)
```

```python
# task 3
class DataSet:
    def __init__(self,size):
        self._size=size
    @property
    def size(self):
        return self._size
    @size.setter
    def size(self,value):
        if value<1:
            raise ValueError("size must be atleast 1")
        self._size=value
obj=DataSet(1000)
print(obj.size)
obj.size=5
print(obj.size)
obj.size=0
```

```python
# task 4 @property
class MLModel:
    def __init__(self, name, accuracy):
        self.name = name
        self.accuracy = accuracy

    @property
    def status(self):
        if self.accuracy >= 0.95:
            return "Production Ready"
        elif self.accuracy >= 0.80:
            return "Needs Improvement"
        else:
            return "Poor"


# Testing
m1 = MLModel("Model A", 0.96)
m2 = MLModel("Model B", 0.85)
m3 = MLModel("Model C", 0.70)

print(m1.status)  # Production Ready
print(m2.status)  # Needs Improvement
print(m3.status)  # Poor
```
