# AI Engineer Python Inheritance

```python
class Animal:
    def __init__(self,name,sound):
        self.name=name
        self.sound=sound
    def speak(self):
        print(f"{self.name} says : {self.sound}")
class Dog(Animal):
    pass
class Cat(Animal):
    pass
dog=Dog("Buddy","woof")
cat=Cat("Kitty","Meow")
dog.speak()
cat.speak()
```

```python
# task 2 method overridding
class MLModel:
    def __init__(self,name,accuracy):
        self.name=name
        self.accuracy=accuracy
    def summary(self):
        print(f" {self.name} has {self.accuracy}")
class DeepLearningModel(MLModel):
    def __init__(self,name,accuracy,layers,framework):
        super().__init__(name,accuracy)
        self.layers=layers
        self.framework=framework
    def summary(self):
        print(f"Name :{self.name}")
        print(f"Accuracy : {self.accuracy*100} ")
        print(f"layers : {self.layers} ")
        print(f"framework :{self.framework}")
dlm=DeepLearningModel("linearRegression",0.92,8,"PyTorch")
dlm.summary()
```

```python
# task 3 polymorphism
class MLModel:
    def evaluate(self):
        print("Evaluation of generic ML model")
class ClassificationModel(MLModel):
    def evaluate(self):
        print("Evaluation of classification model")
class RegressionModel(MLModel):
    def evaluate(self):
        print("Evaluation of regression model")
model1=MLModel()
model2=ClassificationModel()
model3=RegressionModel()
models=[model1,model2,model3]
for model in models:
    model.evaluate()
```

```python
print(isinstance(model2, MLModel))              # True
print(isinstance(model3, ClassificationModel))  # True — because wrong inheritance
```
