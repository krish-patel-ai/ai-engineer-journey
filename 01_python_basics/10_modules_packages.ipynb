```python
# task3 using random module
import random

# Set seed
random.seed(42)

# 5 random floats (0 to 1)
floats = [random.random() for _ in range(5)]
print("Random floats:", floats)

# 3 random integers (1 to 100)
ints = [random.randint(1, 100) for _ in range(3)]
print("Random integers:", ints)
```

```python
# without seed

# Remove seed → results will change every run

floats2 = [random.random() for _ in range(5)]
print("Random floats (no seed):", floats2)

ints2 = [random.randint(1, 100) for _ in range(3)]
print("Random integers (no seed):", ints2)
```

```python
# task4 using the counter
# datetime — dates and times
from datetime import datetime
now = datetime.now()
print(now.strftime("%Y-%m-%d %H:%M:%S"))  # formatted timestamp

# collections — useful data structures
from collections import Counter, defaultdict
words = ["cat", "dog", "cat", "bird", "dog", "cat","fish"]
print(Counter(words))  
most_common = Counter(words).most_common(1)[0]
print(f"Most common: {most_common[0]} appears {most_common[1]} times")
```
