```python
import numpy as np

class SVM:
    def __init__(self, lr=0.001, C=1.0, epochs=1000):
        self.lr = lr
        self.C = C
        self.epochs = epochs
        self.w = None
        self.b = 0

    def fit(self, X, y):
        n, p = X.shape
        self.w = np.zeros(p)
        y = np.where(y <= 0, -1, 1)  # ensure -1/+1 labels

        for _ in range(self.epochs):
            for i in range(n):
                if y[i] * (X[i] @ self.w + self.b) >= 1:
                    self.w -= self.lr * (2 / self.epochs) * self.w
                else:
                    self.w -= self.lr * ((2 / self.epochs) * self.w - self.C * y[i] * X[i])
                    self.b += self.lr * self.C * y[i]

    def predict(self, X):
        return np.sign(X @ self.w + self.b).astype(int)

# --- run it ---
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

X, y = make_classification(n_samples=200, n_features=2, n_redundant=0, random_state=42)
Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=0.2)

svm = SVM()
svm.fit(Xtr, ytr)
print(f"Accuracy: {(svm.predict(Xte) == yte).mean():.2%}")
```
